##**Perceptron(Binary classification)**

```python
import numpy as np
```

Step 1: Prepare Training Data

```python
#Training Data
X=np.array([1,2,3,4])   #Study Hours
y=np.array([0,0,1,1])   #Fail(0)/Pass(1)

X , y
```

Step 2: Initialize Parameters

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("initial Weight:", w)
print("Initial Bias:",b)
```

Define Activation Function(Step Function

```python
def step_function(z):
  return 1 if z>=0 else 0
```

##Training Loop

```python
for epoch in range(epochs):
  print(f"\nEpoch {epoch+1}")
  for i in range(len(X)):
    #1. Compute Weighted sum
    z = w * X[i] + b

    #2.Apply Activation function
    y_pred=step_function(z)

    #3. Compute Error
    error= y[i] - y_pred

    #4.Update Weight
    w = w + learning_rate * error * X[i]

    #5.Update Bias
    b = b + learning_rate * error

    print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, w:{w}, b: {b}")
```

Final model Parameters

```python
print("\nFinal Weight:", w)
print("Final bias:", b)
```

Testing the model

```python
test_value=3.1
z=w*test_value+b
prediction=step_function(z)

print(f"For study hours {test_value}, Prediction", prediction)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X, y=y, hue=y, style=y, s=100, palette='viridis')

# Calculate decision boundary
decision_boundary = -b / w
plt.axvline(x=decision_boundary, color='r', linestyle='--', label=f'Decision Boundary (x={decision_boundary:.2f})')

plt.title('Perceptron Decision Boundary and Training Data')
plt.xlabel('Study Hours (X)')
plt.ylabel('Pass/Fail (0=Fail, 1=Pass)')
plt.yticks([0, 1], ['Fail', 'Pass'])
plt.grid(True)
plt.legend()
plt.show()
```

Perceptron AND Gate

```python
X=np.array([
  [0,0],
  [0,1],
  [1,0],
  [1,1]

])

y=np.array([0,0,0,1])

X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10

print("Initial Weights:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
  return 1 if z>=0 else 0
```

### Training Loop for AND Gate

```python
for epoch in range(epochs):
  print(f"\nEpoch {epoch+1}")
  for i in range(len(X)):
    # 1. Compute Weighted sum
    z = np.dot(w, X[i]) + b

    # 2. Apply Activation function
    y_pred = step_function(z)

    # 3. Compute Error
    error = y[i] - y_pred

    # Only update weights and bias if there's an error
    if error != 0:
      # 4. Update Weights
      w = w + learning_rate * error * X[i]

      # 5. Update Bias
      b = b + learning_rate * error

    print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Weights:{w}, Bias: {b}")
```

### Final Model Parameters for AND Gate

```python
print("\nFinal Weights:", w)
print("Final Bias:", b)
```

### Testing the AND Gate Perceptron

```python
print("Testing the trained AND gate perceptron:")
for i in range(len(X)):
  test_input = X[i]
  predicted_output = step_function(np.dot(w, test_input) + b)
  actual_output = y[i]
  print(f"Input: {test_input}, Predicted: {predicted_output}, Actual: {actual_output}")
```

## Perceptron for OR Gate
